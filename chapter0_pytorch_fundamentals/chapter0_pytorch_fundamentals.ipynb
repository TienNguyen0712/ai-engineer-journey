### **Nhân ma trận**

Hai cách chính để niểu diễn nhận ma trận trong mạng nơ ron và học sâu:

1. Nhân các phần tử:  Nhân từng phần tử trong ma trận với từng phần tử trong ma trận kia
2. Nhân ma trận: Phẩn từ dòng thứ i sẽ nhân với phần tử cột j để hình thành phần tử dòng i cột j sau khi cộng lại. Đặc biệt là số hàng ma trận này phải bằng số cột mâ trận kia  (dùng dấu .)

Thông tin thêm: https://www.mathsisfun.com/algebra/matrix-multiplying.html

Hai lỗi phổ biến khi nhân ma trận

1. Các ma trận phải chung chiều với nhau tức số hàng ma trận này là số cột ma trận kia:
- `(3, 2) @ (3, 2)` không đúng
- `(2, 3) @ (3, 2)` đúng
- `(3, 2) @ (2, 3)` đúng

2. Chiều của ma trận kết quả sẽ là shape sau khi loại bỏ 2 hàng và cột trùng

- `(2, 3) @ (3, 2)` -> `(2, 2)`
- `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
# Nhân từng phần tử
tensor = torch.tensor([1, 2, 3])
print(tensor, "*", tensor)
print(f"Kết quả: {tensor * tensor}")
```

### **Lỗi phổ biến trong Học sâu là sai shape**

```python
# Shape khi nhân ma trân
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])


torch.mm(tensor_A, tensor_B) # torch.mm là torch.matmul


# Lỗi chung shape
```


Để fix tensor shape ta cần điều chỉnh 1 trong 2 shape của ma trận sử dụng **transpose**

**Transpose** chuyển cột thạnh hàng và thành hàng thành cột

```python
tensor_B.T, tensor_B.T.shape
```

```python
# phếp nhân ma trận này hoạt động khi ma trận B chuyển vị
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} chung shape trên, tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")

output = torch.mm(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

### **Đổi chiều chuyển thành stacking, squeezing và unsqueezing**

Đôi khi ta muốn thay đổi hàng và cột hoặc thay đổi chiều của tensor mà không thay đổi các giá trị trong chúng. Ta sử dụng các phương pháp sau:

- `torch.reshape(input, shape)`: để đổi shape nếu không có liên tục tự động copy dữ liệu   
- `torch.view(shape)`: trả về view mới đối với tensor trong nhiều shape khác nhau nhưng với dữ liệu giống với tensor gốc
- `torch.stack(tensors, dim=0)`: Chồng các tensor lên nhau và tạo ra một chiều mới. Các tensor phải có cùng kích thước.
- `torch.squeeze(input)`: Loại bỏ chiều có kích thước là 1
- `torch.unsqueeze(input)`: Thêm chiều mới có kích thước là 1
- `torch.permute(intput, dims)`: Dùng để hoán đổi thứ tự các chiều không gian.

```python
# Dùng reshape
x = torch.arange(1., 8.)
x, x.shape
```

```python
# Thêm một chiều
x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape
```

```python
# Change view (keeps same data as original but changes view)
# See more: https://stackoverflow.com/a/54507446/7900723
z = x.view(1, 7)
z, z.shape
```

```python
z[:, 0] = 5
z, x
```

```python
# Xếp tensor lên những tensor còn lại
x_stacked = torch.stack([x, x, x, x], dim=0) # 0: theo hàng, 1: theo cột
x_stacked
```

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Loại bỏ chiều vừa thêm trong tensor x_squeezed
x_squeezed = x_reshaped.squeeze() # Dùng squeeze
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Thêm một chiều với unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# Tạo tensor ban đầu
x_original = torch.rand(size=(224, 224, 3))

# Điều chỉnh shape thành các vị trí khác nhau
x_permuted = x_original.permute(2, 0, 1) # chuyển axis 0->2, 1->0, 2->1

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

**Lưu ý**

Cả `permute` và `view` đều chạy trên dữ liệu chính do đó mà khi thay đổi các dữ liệu trong tensor tạo ra bởi chùng thì sẽ ảnh hưởng đến tensor gốc

### **Pytorch tensors & Numpy**

Hai phương thức chính để sử dụng Numpy chung với Pytorch là:

- `torch.from_numpy(ndarray)`: Chuyển Numpy array thành Pytorch tensor

- `torch.Tensor.numpy()`: Chuyển Pytorch tensor thành Numpy array


```python
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

**Lưu ý**

Mặc định Numpy array là `float64` và nếu chuyển thành tensor sẽ giữ nguyên kiểu dữ liệu

Tuy nhiên một số Pytorch sử dụng `float32` do đó khi chuyển muốn thực hiện tính toán ta cần chuyển về đúng kiểu dữ liệu


```python
# Thay đổi array giữ lại tensor
array = array + 1
array, tensor
```

```python
# Chuyển tensor thành Numpy array
tensor = torch.ones(7) # Tạo tensor với kiểu dữ liệu float32
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Thay đổi tensor vè giữ nguyên array
tensor = tensor + 1
tensor, numpy_tensor # Không tạo bản sao
```

### **Tái sử dụng**

Khi học về mạng nơ ron và học máy ta sẽ bắt đầu khám phá rằng tính ngẫu nhiên của phép toán này là bao nhiêu

Ta biết rằng khi mạng nơ ron học bằng cách bắt dầu với con số ngẫu nhiên để mô tả một thiết kế của dữ liệu và cố gắng cải thiện con số ngẫu nhiên đó sử dụng các phép tính quan hệ tensor. Nói ngắn gọn

> Bắt đầu với số ngẫu nhiên -< Phép quan hệ tensor -> Cái thiện

Tuy nhiên không phải ngẫu nhiên lúc nào cũng cần thiết. Lấy ví dụ khi viết một thuật toán và so sánh nhiều thuật toán khác nhau trên cũng một dự liệu ngẫu nhiên nào đó. Đó chính là lúc việc tái sử dụng xuất hiện


```python

import torch

# Tạo 2 tensor ngẫu nhiên
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(f"Tensor A:\n{random_tensor_A}\n")
print(f"Tensor B:\n{random_tensor_B}\n")
print(f"Does Tensor A equal Tensor B? (anywhere)")
random_tensor_A == random_tensor_B
```

Ta sử dụng `torch.manual.seed(seed)` để thực hiện giữ nguyên các giá trị ngẫu nhiên

```python
import torch
import random

# Set seed
RANDOM_SEED=42 # Giá trị seed
torch.manual_seed(seed=RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

# reset seed sau mỗi lẫn hàm rand() được gọi
# Nếu không có điều này, tensor_D sẽ khác giá trị to tensor_C
torch.random.manual_seed(seed=RANDOM_SEED) # phải set trước để cùng giá trị
random_tensor_D = torch.rand(3, 4)

print(f"Tensor C:\n{random_tensor_C}\n")
print(f"Tensor D:\n{random_tensor_D}\n")
print(f"Does Tensor C equal Tensor D? (anywhere)")
random_tensor_C == random_tensor_D
```
